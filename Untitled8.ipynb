```python
!pip install transformers
!pip install torch
import pandas as pd

data2=pd.read_csv('/content/final.csv')
data3=data2['0']
data4=data2.drop("0.000123, which corresponds to a distance of 705 Mly, or 216 Mpc.", axis='columns')

from transformers import GPT2Tokenizer, GPT2LMHeadModel, TextDataset, DataCollatorForLanguageModeling, Trainer, TrainingArguments

tokenizer = GPT2Tokenizer.from_pretrained('gpt2')

tokenizer = GPT2Tokenizer.from_pretrained('gpt2')

tokenizer.add_special_tokens({'pad_token': '[PAD]'})

tokenizer.pad_token = tokenizer.eos_token




max_length = 50  # specify the desired maximum length for your sequences

tokenized_datas = []

for dt in data4:

    tokenized = tokenizer.encode(dt, padding='max_length',truncation=True, max_length=max_length)

    tokenized_datas.append(tokenized)


import torch

tensor_text = torch.tensor(tokenized_datas).unsqueeze(0)


batch_size = 8

num_epochs = 10

learning_rate = 5e-5

from torch.utils.data import DataLoader, TensorDataset

dataset = TensorDataset(tensor_text)

dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

model = GPT2LMHeadModel.from_pretrained("gpt2")

model.train()


model = GPT2LMHeadModel.from_pretrained("gpt2")


optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)


for epoch in range(num_epochs):

    for batch in dataloader:

        optimizer.zero_grad()

        input_ids = batch[0].to(model.device)

        outputs = model(input_ids, labels=input_ids)

        loss = outputs[0]

        loss.backward()

        optimizer.step()



from transformers import pipeline



tokenizer = GPT2Tokenizer.from_pretrained('gpt2')



# Set the model to evaluation mode

model.eval()



# Generate text using the model

text_generator = pipeline('text-generation', model=model, tokenizer=tokenizer)

generated_text = text_generator("the goverment is", max_length=150, do_sample=True)[0]['generated_text']



print(generated_text)

encoded_data = tokenizer.batch_encode_plus(

    data4,

    add_special_tokens=True,

    return_attention_mask=True,

    return_tensors='tf'

)
```
